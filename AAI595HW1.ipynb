```python
order = []
mse = []
def polynomial_fit(m):
    plt.plot(x,y,'bo')
    coefficients = np.polyfit(np.reshape(x,(1,50))[0],np.reshape(y,(1,50))[0],deg = m)
    poly_function = np.poly1d(coefficients)
    x1 = np.linspace(min(x), max(x), len(x))
    y1 = poly_function(x1)
    y_pred = poly_function(x)
    msesh = mean_squared_error(y,y_pred) 
    order.append(m)
    mse.append(msesh)
    plt.plot(x1,y1,"c")
    plt.title("M = " + str(m))
    plt.show()
```

Based on the above graphs The best choice of m would be degree 3. Degree 3 is the elbow point of the graph where even though the following degrees have lower mse values it is of diminishing returns. With each increase of degree from 3 mse decreases slightly but computational cost increases as well as the risk of overfitting the dataset so with that being considered order 3 is ideal.

```python
m = 3
noise = []
mse = []
def generate_noise(ns):
    noise_scale = ns
    number_of_samples = 50
    x = 30*(np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    plt.plot(x,y,'bo')
    coefficients = np.polyfit(np.reshape(x,(1,50))[0],np.reshape(y,(1,50))[0],deg = m)
    poly_function = np.poly1d(coefficients)
    x1 = np.linspace(min(x), max(x), len(x))
    y1 = poly_function(x1)
    y_pred = poly_function(x)
    msesh = mean_squared_error(y,y_pred)
    plt.plot(x1,y1,"c")
    plt.title("Noise Scale = " + str(noise_scale))
    plt.show()
    noise.append(noise_scale)
    mse.append(msesh)
```

Here it is shown in this data that increasing noise_scale results in higher MSE. Therefore we would want a lower noise scale if we cannot change the order or other parameters. Dealing with noise scale can be done by cleaning the data and preprocessing it before using it in a model.

```python
samples = []
mse = []
m = 3
noise_scale = 100 # back to default value
def generate_samples(sample):
    number_of_samples = sample
    x = 30*(np.random.rand(number_of_samples, 1) - 0.5)
    y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
    plt.plot(x,y,'bo')
    coefficients = np.polyfit(np.reshape(x,(1,number_of_samples))[0],np.reshape(y,(1,number_of_samples))[0],deg = m)
    poly_function = np.poly1d(coefficients)
    x1 = np.linspace(min(x), max(x), len(x))
    y1 = poly_function(x1)
    y_pred = poly_function(x)
    msesh = mean_squared_error(y,y_pred)
    plt.plot(x1,y1,"c")
    plt.title("Number of Samples = " + str(number_of_samples))
    plt.show()
    samples.append(number_of_samples)
    mse.append(msesh)
```

```python
generate_samples(40)
```

```python
generate_samples(30)
```

```python
generate_samples(20)
```

```python
generate_samples(10)
```

```python
plt.plot(samples,mse)
plt.title("Number of Samples vs MSE")
```

As number of samples increase the mse decreases. This is because there is more information avaliable for the polyfit function to corrrect the parameters. It also means that outliers have less of an impact on the line of best fit since there are more
samples to mitigate it. However when there are less samples, the effect of outliers increase since it holds more weight since there are less samples to balance it out. When building a model it is often worthwhile to increase the amount of datapoints. Sometimes though it may lead to overfitting based on the type of sample data with reference to the population.
